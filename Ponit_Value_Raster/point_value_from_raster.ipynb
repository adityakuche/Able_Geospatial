```python
%matplotlib inline
import matplotlib.pyplot as plt
import geopandas as gpd
import rasterio
from rasterio.plot import show
```

```python
#importing shapefile from the system
data = gpd.read_file('point_1.shp')
print(data.crs)
data.plot()
```

```python
#importing raster
ras = rasterio.open('ortho.tif')
print(ras.crs)
print(ras.count)
```

```python
#point and raster
fig, ax = plt.subplots(figsize=(10,10))
data.plot(ax=ax, color= 'red')
show(ras,ax=ax)
```

```python
#extract xy from  point
for p in data['geometry']:
    print(p.xy[0][0],p.xy[1][0])
    
```

```python
#extract point value from raster
for p in data['geometry']:
    x= p.xy[0][0]
    y= p.xy[1][0]
    row, col = ras.index(x,y)
    print('row, col: %d, %d'%(row, col))
    print('point_value %.2f \n'%ras.read(1)[row,col])
```
