```python
import pandas as pd
import numpy as np
from geopandas. tools import geocode 
import geopandas as gpd
```

```python
df= pd.read_csv("List of forests in India.csv") #import the csv or excel as required(to_excel)
```

```python
df.columns #checking up the columns their in the dataset
```

```python
df=df.drop(columns=['lon', 'lat','Unnamed: 0'],axis=1) #axis=1 referse to columns and 0 refers to row, removing the unwanted columns
```

# Geocoding from Address/Name columns to Latitude/Longitude columns

```python
df
```

```python
#Calculating lat/long according to the name of national parks in the row
for index, row in df.iterrows():
    print (row['Name'])
    
    info= geocode(str(row['Name']), provider='arcgis',timeout=10) #nominator can be use but in my case it is not working so, I use arcgis provider
    #df.loc[int(index), 'Address']= info['address'].loc[0]
    df.loc[int(index), 'lon']=info['geometry'].loc[0].x
    df.loc[int(index), 'lat']=info['geometry'].loc[0].y
```

```python
df
#df.to_csv("Name_geocoded.csv") #This can save this geocode data to the 
```

# Converting csv_to_shapefile

```python
# To save the csv file to shapefile
# We now create a geodataframe from our data.
data_gdf = gpd.GeoDataFrame(df, geometry = gpd.points_from_xy(df['lon'], df['lat']))
```

```python
data_gdf.info() # Here datatype is geodataframes
# type(data_gdf) #to get the datatype of variable
```

```python
#ESRI_WKT = 'GEOGCS["GCS_WGS_1984",DATUM["D_WGS_1984",SPHEROID["WGS_1984",6378137,298.257223563]],PRIMEM["Greenwich",0],UNIT["Degree",0.017453292519943295]]'
data_gdf.to_file(filename='data.shp', driver = 'ESRI Shapefile', crs="EPSG:4326") #EPSG:4326
# you can also get the ESRI_WKT to the crs instead of 'EPSG:4326' whichever is the suitable
```
